```python
import matplotlib.pyplot as plt
x=[2023, 2024, 2025, 2026]
y=[15, 25, 100, 88]
plt.plot(x, y)
plt.show()
```

```python
import numpy as np
x=np.array([2023, 2024, 2025, 2026])
y=np.array([15, 25, 100, 88])
plt.plot(x, y)
plt.title("Matplotlib Test")
plt.xlabel("Year")
plt.ylabel("Values")
plt.show()
```
