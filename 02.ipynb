Matplotlib documentation:   https://matplotlib.org/stable/

List of matplotlib markers that can be used:  https://matplotlib.org/stable/api/markers_api.html

```python
# implementing markers in line plot
import matplotlib.pyplot as plt
import numpy as np

x=np.array([2023, 2024, 2025, 2026])
y=np.array([15, 25, 100, 88])
plt.plot(x, y, marker='*')
plt.show()
```

```python
# markerSize

x=np.array([2023, 2024, 2025, 2026])
y=np.array([15, 25, 100, 88])
plt.plot(x, y, marker='*', markersize=10)
plt.show()
```

```python
# markerFaceColor

x=np.array([2023, 2024, 2025, 2026])
y=np.array([15, 25, 100, 88])
plt.plot(x, y, marker='*', markersize=10, markerfacecolor='#1cd3fc')
plt.show()
```

```python
# markerEdgeColor

x=np.array([2023, 2024, 2025, 2026])
y=np.array([15, 25, 100, 88])
plt.plot(x, y, marker='*', markersize=10, markerfacecolor="#04323c", markeredgecolor='blue')
plt.show()
```

```python
# markerEdgeWidth width

x=np.array([2023, 2024, 2025, 2026])
y=np.array([15, 25, 100, 88])
plt.plot(x, y, marker='*', markersize=10, markerfacecolor="#04323c", markeredgecolor='blue', markeredgewidth=5)
plt.show()
```

```python
# lineStyle (dashed, dotted, dashdot, solid)
x=np.array([2023, 2024, 2025, 2026])
y=np.array([15, 25, 100, 88])
plt.plot(x, y, marker='*', markersize=10, markerfacecolor="#04323c", markeredgecolor='blue', markeredgewidth=3, linestyle='--')
plt.show()
```

```python
# lineStyle (dashed, dotted, dashdot, solid, none)
x=np.array([2023, 2024, 2025, 2026])
y=np.array([15, 25, 100, 88])
plt.plot(x, y, marker='*', markersize=10, markerfacecolor="#04323c", markeredgecolor='blue', markeredgewidth=3, linestyle='solid')
plt.show()
```

```python
# lineWidth
x=np.array([2023, 2024, 2025, 2026])
y=np.array([15, 25, 100, 88])
plt.plot(x, y, marker='*', markersize=10, markerfacecolor="#04323c", markeredgecolor='blue', markeredgewidth=3, linestyle='solid', linewidth=5)
plt.show()
```

```python
# lineColor
x=np.array([2023, 2024, 2025, 2026])
y=np.array([15, 25, 100, 88])
plt.plot(x, y, marker='*', markersize=10, markerfacecolor="#04323c", markeredgecolor='blue', markeredgewidth=3, linestyle='solid', linewidth=5, color='purple')
plt.show()
```

```python
# multipleLines

x=np.array([2023, 2024, 2025, 2026])
y1=np.array([15, 25, 100, 88])
y2=np.array([17, 23, 38, 5])

plt.plot(x, y1, marker='*', markersize=10, markerfacecolor="#04323c", markeredgecolor='blue', markeredgewidth=3, linestyle='solid', linewidth=5, color='purple')
plt.plot(x, y2)
plt.show()
```

```python
# multipleLines using arguments inside a dictionary (use **line_style to unpack the dictionary)

line_style=dict(marker='*', markersize=10, markerfacecolor="#04323c", markeredgecolor='blue', markeredgewidth=3, linestyle='solid', linewidth=5, color='purple')

plt.plot(x, y1, **line_style)
plt.plot(x, y2, **line_style)

plt.show()
```

```python
# multipleLines (3 lines from different data sets)

x=np.array([2023, 2024, 2025, 2026])
y1=np.array([15, 25, 100, 88])
y2=np.array([17, 23, 38, 5])
y3=np.array([13, 21, 33, 24])

plt.plot(x, y1, **line_style)
plt.plot(x, y2, **line_style)
plt.plot(x, y3, **line_style)

plt.show()
```

```python
# multipleLines, different line color for each line, change the arguments inside the dictionary

line_style_changed=dict(marker='*', markersize=10, markerfacecolor="#04323c", markeredgecolor='blue', markeredgewidth=3, linestyle='solid', linewidth=5,)

plt.plot(x, y1, **line_style_changed, color='purple')
plt.plot(x, y2, **line_style_changed, color='green')
plt.plot(x, y3, **line_style_changed, color='orange')

plt.show()
```
