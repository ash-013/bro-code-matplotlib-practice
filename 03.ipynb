```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.array([2023, 2024, 2025, 2026])
y1=np.array([15, 25, 100, 88])
y2=np.array([17, 23, 38, 5])
y3=np.array([13, 21, 33, 24])

plt.title("Class Size", fontsize=20, family='serif', fontweight='bold', color='green', pad=10)

plt.xlabel("Year", fontsize=15, family='serif', fontweight='bold', color='blue')

plt.ylabel("Number of Students", fontsize=15, family='serif', fontweight='bold', color='blue')

plt.tick_params(axis='both', colors="#000000") 
plt.xticks(x)
plt.subplots_adjust(bottom=0.175, left=0.14, right=0.95)
plt.plot(x, y1)
plt.plot(x, y2)
plt.plot(x, y3)
plt.show()
```
