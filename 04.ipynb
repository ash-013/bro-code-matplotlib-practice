```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Grid add grid lines to the plot 
# linestyle include: solid, dashed, dashdot, dotted
# axis include: x, y, both
# linewidth: line width
# color: line color

x=[1,2,3,4,5]
y=[2,3,5,7,11]
plt.plot(x,y)
plt.grid(axis='x', linewidth=2, color='lightgray', linestyle='dotted')
plt.show()
```

```python
x=[1,2,3,4,5]
y=[2,3,5,7,11]
plt.plot(x,y)
plt.grid(axis='x', linewidth=2, color='lightgray', linestyle='dotted')
plt.grid(axis='y', linewidth=2, color='lightgray', linestyle='solid')
plt.show()
```

```python
x=[1,2,3,4,5]
y=[2,3,5,7,11]
plt.plot(x,y)
plt.grid(axis='x', linewidth=2, color='lightgray', linestyle='dashed')
plt.grid(axis='y', linewidth=2, color='lightgray', linestyle='dotted')
plt.show()
```

```python

x=[1,2,3,4,5]
y=[2,3,5,7,11]
plt.plot(x,y)
plt.grid(axis='both', linewidth=2, color='lightgray', linestyle='dotted')
plt.show()
```
