```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# bar-chart -> compare categories of the data

categories=np.array(['Grains', 'Fruit', 'Vegetables', 'Dairy', 'Salad', 'Sandwiches'])
values=np.array([5, 3, 4, 2, 6, 4])

plt.bar(categories, values, color='skyblue', edgecolor='blue', width=0.8, alpha=0.8)

plt.xlabel('Food', fontsize=12)
plt.ylabel('Quantity', fontsize=12)
plt.title('Daily Consumption')
plt.show()
```

```python
# horizontal bar-chart (use barh instead of bar)

categories=np.array(['Grains', 'Fruit', 'Vegetables', 'Dairy', 'Salad', 'Sandwiches'])
values=np.array([5, 3, 4, 2, 6, 4])

plt.barh(categories, values, color='gray', edgecolor='blue', height=0.8, alpha=0.8)

plt.xlabel('Quantity', fontsize=12)
plt.ylabel('Food', fontsize=12)
plt.title('Daily Consumption')
plt.show()
```
