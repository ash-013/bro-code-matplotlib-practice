```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# basic pie chart

categories=["Freshman","Sophomore","Juniors","Seniors"]
values=[100,250,220,140]

plt.pie(values)
plt.show()
```

```python
# basic pie chart with labels

plt.pie(values,labels=categories)
plt.show()
```

```python
# basic pie chart with labels and percentage

plt.pie(values,labels=categories,autopct="%1.1f%%")
plt.show()
```

```python
# basic pie chart with labels, percentage and custom colors

categories=["Freshman","Sophomore","Juniors","Seniors"]
values=[100,80,60,40]
colors=["red","yellow","blue","green"]

plt.pie(values,labels=categories,autopct="%1.1f%%",colors=colors)
plt.title("Class Distribution")
plt.show()
```

```python
# Explode the a custom slice (Seniors)

categories=["Freshman","Sophomore","Juniors","Seniors"]
values=[100,80,60,40]
colors=["red","yellow","blue","green"]
explode = (0, 0, 0, 0.1)  # only "explode" the 4th slice (Seniors)

plt.pie(values,labels=categories,autopct="%1.1f%%",colors=colors, explode=explode)
plt.title("Class Distribution")
plt.show()
```

```python
# include the shadow

categories=["Freshman","Sophomore","Juniors","Seniors"]
values=[100,80,60,40]
colors=["red","yellow","blue","green"]
explode = (0, 0, 0, 0.1)  # only "explode" the 4th slice (Seniors)

plt.pie(values,labels=categories,autopct="%1.1f%%",colors=colors, explode=explode, shadow=True)
plt.title("Class Distribution")
plt.show()
```

```python
# rotate the start angle

categories=["Freshman","Sophomore","Juniors","Seniors"]
values=[100,80,60,40]
colors=["red","yellow","blue","green"]
explode = (0, 0, 0, 0.1)  # only "explode" the 4th slice (Seniors)

plt.pie(values,labels=categories,autopct="%1.1f%%",colors=colors, explode=explode, shadow=True, startangle=90)
plt.title("Class Distribution")
plt.show()
```
