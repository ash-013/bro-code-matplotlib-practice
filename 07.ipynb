```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# basic scatter plot (scatter plots show relationships between two variables, helps identify correlations (positive/negative/no correlation), clusters, and outliers)

x=[0, 1, 1, 2, 3, 4, 5, 6, 7, 7, 8]  # x-> Study hours
y=[55, 60, 62, 59, 65, 70, 75, 80, 85, 88, 95] # y-> Scores

plt.scatter(x,y)
plt.show()
```

```python
x=[0, 1, 1, 2, 3, 4, 5, 6, 7, 7, 8] 
y=[55, 60, 62, 59, 65, 70, 75, 80, 85, 88, 95]

plt.scatter(x,y)
plt.xlabel("Study Hours")
plt.ylabel("Scores")
plt.title("Study Hours vs Scores")
plt.show()
```

```python
# color,size and alpha

x=[0, 1, 1, 2, 3, 4, 5, 6, 7, 7, 8] 
y=[55, 60, 62, 59, 65, 70, 75, 80, 85, 88, 95]

plt.scatter(x,y ,color='green', s=100, alpha=0.5)
plt.xlabel("Study Hours")
plt.ylabel("Scores")
plt.title("Study Hours vs Scores")
plt.show()
```

```python
# multiple classes in scatter plot with labels, legend and different colors 

x1=np.array([0, 1, 1, 2, 3, 4, 5, 6, 7, 7, 8] )
y1=np.array([55, 60, 62, 59, 65, 70, 75, 80, 85, 88, 95])

x2=np.array([0, 1, 2, 2, 3, 4, 5, 6, 7, 8, 8] )
y2=np.array([50, 58, 62, 70, 72, 78, 83, 88, 92, 98, 97])

plt.scatter(x1,y1)
plt.scatter(x1,y1, color='skyblue', alpha=0.5, s=100, label='Class A')
plt.scatter(x2,y2, color='red', alpha=0.5, s=100, label='Class B')
plt.xlabel("Study Hours")
plt.ylabel("Scores")
plt.title("Study Hours vs Scores")
plt.legend()
plt.show()
```
