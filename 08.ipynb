```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Histograms are the visual representation of the distribution of quantitative data. Histograms group data into bins, displaying the frequency of data points within each bin.

scores=np.random.normal(loc=70, scale=10, size=1000)
plt.hist(scores)
plt.show()
```

```python
# Clip function can be used to limit the values in an array. No score below the lower limit and no score above the upper limit

scores=np.random.normal(loc=70, scale=10, size=1000)
scores=np.clip(scores, 0, 100)
plt.hist(scores)
plt.show()
```

```python
# bins can be used to specify the number of intervals in the histogram

scores=np.random.normal(loc=70, scale=10, size=1000)
scores=np.clip(scores, 0, 100)
plt.hist(scores, bins=20)
plt.show()
```

```python
# changing the color and edge color of the histogram
scores=np.random.normal(loc=70, scale=10, size=1000)
scores=np.clip(scores, 0, 100)
plt.hist(scores, color='lightgreen', bins=20, edgecolor='black')
plt.title('Distribution of Exam Scores')
plt.xlabel('Scores')
plt.ylabel('Number of Students')
plt.show()
```
