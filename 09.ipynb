```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Figure -> The entire canvas
# Ax -> A single plot (subplot)

print(plt.subplots(2,2))
```

```python
# plotting on a specific ax, here axes[0,0]

x=np.array([1,2,3,4,5])
figure, axes = plt.subplots(2,2)

axes[0,0].plot(x,x**2, color='red')
axes[0,0].set_title('y=x^2')
axes[0,0].set_xlabel('x')
axes[0,0].set_ylabel('y')

axes[0,1].plot(x,x**3, color='green')
axes[0,1].set_title('y=x^3')
axes[0,1].set_xlabel('x')
axes[0,1].set_ylabel('y')

axes[1,0].plot(x, x**3-3*(x**2)+2*x-88, color='purple')
axes[1,0].set_title('y=x^3-30x^2+2x-88')
axes[1,0].set_xlabel('x')
axes[1,0].set_ylabel('y')

axes[1,1].plot(x, -x**2, color='pink')
axes[1,1].set_title('y=-x^2')
axes[1,1].set_xlabel('x')
axes[1,1].set_ylabel('y')

plt.tight_layout()
plt.show()
```
