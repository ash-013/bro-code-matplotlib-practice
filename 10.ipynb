```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('dataset_for_file_10.csv')
df.head()
```

```python
type_count=df['Type1'].value_counts(ascending=True)
plt.barh(type_count.index, type_count.values, color='gray', edgecolor='black', linewidth=1)
plt.title('Number of Pokemon by primary types')
plt.xlabel('Count')
plt.ylabel('Type1')
plt.tight_layout()
plt.show()
```
